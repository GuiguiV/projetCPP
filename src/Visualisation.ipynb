```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation
```

```python
with open("out_expl.txt") as file:
    lines = file.readlines()
    N_etapes = len(lines)
    array = [line.split(" ")[:-1] for line in lines]
    taille_portion = len(array[0])
    heat = np.array(array).astype(np.float)
    print(heat.shape)
```

```python
%matplotlib notebook
fig, ax = plt.subplots()
line, = ax.plot(heat[0])
ax.set_ylabel("T")
ax.set_xlabel("x")


def animate(t):
    line.set_ydata(heat[t])
    ax.set_title("Euler explicite : t : "+str(t))
    return line,
ani = animation.FuncAnimation(fig, animate, interval=20, blit=True, repeat=True, save_count=50)
plt.show()
```
